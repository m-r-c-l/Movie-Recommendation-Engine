```python
import dask.dataframe as dd
from imdb import IMDb
from dask.diagnostics import ProgressBar
from dask.distributed import Client, LocalCluster
from tqdm import tqdm

# Step 1: Set up a Dask LocalCluster to utilize all available RAM
# Configure the cluster to use 24GB of memory, divided across workers
cluster = LocalCluster(n_workers=4, threads_per_worker=2, memory_limit='6GB')  # Adjust based on your 24GB RAM system
client = Client(cluster)

# Display the Dask dashboard link to monitor progress
client

# Step 2: Define a list of specific IMDb IDs to process
imdb_id_list = ['0133093', '0111161', '0068646', '0468569']  # Add IMDb IDs you want to process

# Step 2: Load your dataset using Dask
file_path = '/Users/timjamboula/Documents/Movie-Recommendation-Engine/raw_data/ml-latest-small/movie_imdb_cleaned_final.csv'
ddf = dd.read_csv(file_path)

# Convert 'imdbId' to string and strip whitespace in parallel
ddf['imdbId'] = ddf['imdbId'].astype(str).str.strip()

# Ensure 'imdbId' only contains numeric characters
ddf['imdbId'] = ddf['imdbId'].str.extract(r'(\d+)', expand=False)

# Drop duplicates in parallel
ddf = ddf.drop_duplicates(subset='imdbId', keep='first')

# Initialize and activate the Dask progress bar
pbar = ProgressBar()
pbar.register()

# Step 3: Compute the Dask DataFrame into a Pandas DataFrame (with progress bar)
original_own_df = ddf.compute()

# Step 4: IMDb Poster Scraping
# Initialize IMDbPY object
ia = IMDb()

# Function to fetch movie poster using IMDbID
def fetch_movie_poster(imdb_id):
    try:
        # Search movie using IMDbID
        movie = ia.get_movie(imdb_id)
        # Return the poster URL if it exists
        return movie['cover url'] if 'cover url' in movie else None
    except Exception as e:
        print(f"Error fetching poster for IMDb ID {imdb_id}: {e}")
        return None

# Step 5: Fetch movie posters using tqdm progress bar for IMDb scraping
tqdm.pandas()
original_own_df['movie_poster'] = original_own_df['imdbId'].progress_apply(lambda x: fetch_movie_poster(str(x).zfill(7)))

# Step 6: Save the cleaned DataFrame with movie poster URLs to a new CSV
output_file_path = '/Users/timjamboula/Documents/Movie-Recommendation-Engine/raw_data/ml-latest-small/imdb_movie_poster.csv'
original_own_df.to_csv(output_file_path, index=False)

# Display the first few rows to verify
print(original_own_df.head())
```
